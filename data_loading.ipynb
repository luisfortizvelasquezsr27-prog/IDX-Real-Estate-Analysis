# Week 1 – Data Loading & Aggregation

Reads all monthly CRMLS CSV files, concatenates them into two combined datasets (listings and sold), filters to `PropertyType == 'Residential'`, and saves the results for use in the analysis notebook.

```python
import pandas as pd
import glob

# Discover monthly files
listing_files = sorted(glob.glob("CRMLSListing*.csv"))
sold_files    = sorted(glob.glob("CRMLSSold*.csv"))

print("Listing files found:", listing_files)
print("Sold files found:   ", sold_files)
```

```python
# Concatenate all monthly listing files
listings_df = pd.concat(
    [pd.read_csv(f, low_memory=False) for f in listing_files],
    ignore_index=True
)

# Concatenate all monthly sold files
sold_df = pd.concat(
    [pd.read_csv(f, low_memory=False) for f in sold_files],
    ignore_index=True
)

print(f"Listings rows (before filter): {len(listings_df):,}")
print(f"Sold rows     (before filter): {len(sold_df):,}")

display(listings_df.head())
display(sold_df.head())
```

```python
# Filter to Residential only
listings_rows_before = len(listings_df)
sold_rows_before     = len(sold_df)

listings_res = listings_df[listings_df['PropertyType'] == 'Residential'].copy()
sold_res     = sold_df[sold_df['PropertyType'] == 'Residential'].copy()

print("LISTINGS:")
print(f"  Before filter: {listings_rows_before:,} rows")
print(f"  After  filter: {len(listings_res):,} rows")
print(f"  Dropped:       {listings_rows_before - len(listings_res):,} rows")

print("\nSOLD:")
print(f"  Before filter: {sold_rows_before:,} rows")
print(f"  After  filter: {len(sold_res):,} rows")
print(f"  Dropped:       {sold_rows_before - len(sold_res):,} rows")
```

```python
# Save Residential-filtered datasets for the analysis notebook
listings_res.to_csv('listings_residential.csv', index=False)
sold_res.to_csv('sold_residential.csv', index=False)

print("Saved:")
print(f"  listings_residential.csv  – {len(listings_res):,} rows")
print(f"  sold_residential.csv      – {len(sold_res):,} rows")
```
